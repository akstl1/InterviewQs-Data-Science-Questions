```python
import random
import numpy
import seaborn as sns
import plotly.express as px
```

```python
def dice():
    roll1 = random.randint(1, 6)
    roll2 = random.randint(1, 6)
    total = roll1+roll2
    return total
```

```python
rolls = []
average=[]
for i in range(500):
    val = dice()
    rolls.append(val)
    average.append(numpy.mean(rolls))
    fig = px.line(average)
fig.show()
```
